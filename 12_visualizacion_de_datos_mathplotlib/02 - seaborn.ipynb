# Parametros mas usados de seaborn

Seaborn tiene una gran cantidad de graficos, y una gran cantidad de parametros similares en cada uno de los graficos

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# histograma
sns.displot(data=tips, x="total_bill")
plt.show()
```

```python
# tip (propina)
# total _bil (total de cuenta)
# se incluye datos en 'y'
sns.displot(data=tips, x="total_bill", y='tip')
plt.show()
```

```python
# estos los parametros mas importantes
# data, valores de (x,y), hue (segmentos)
sns.displot(data=tips, x="total_bill", y='tip', hue='sex')
plt.show()
```

```python
# solo valores de x
sns.displot(data=tips, x="total_bill", hue='sex')
plt.show()
```

```python
# data
# valores de (x,y)
# hue (segmentos)
# kind (tipo de graf)
# legend (leyenda)
# alpha (transparencia)

sns.displot(data=tips, x="total_bill", hue='sex', kind='kde', legend=True, palette='dark', alpha=.7)
plt.show()
```

# Distribuaciones en Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# grafico de histograma

# stat (tipo de conteo se quiere):
    # count (default)
    # frequency
    # probability
    # percent
    # density
    
# multiple (forma de presentacion):
    # stack
    # layer (default)
    # dodge
    # fill 
sns.histplot(data=tips, x='tip', bins=15, hue='sex', stat='count', multiple='fill')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True)
plt.show()
```

```python
# diagram de densidad,
# muestra el comportamiento de la propinas

# fill (pinta el area bajo la curva)
# bw_adjust (ajusta la grafica a 1)
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, fill=True, bw_adjust=1)
plt.show()
```

```python
# grafico escalonado que tienen las propinas a lo largo de su frecuencia
# se lo puede presenta de varias maneras con 'stat'
    # count
    # proportion
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
# sin importar la funcion de distribucion que se escoja, 'kind' lo puede modificar a otro tipo
# se puede usar cualquiera de la graficos de distribucion, 'kind' lo cambia sin problemas
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show()
```
