# Metodo orientado a objetos

Existen diferentes formas de ejecutar graficas o vizualizaciones en Mathplotlib
* Uno de estos metodos es **Pyplot**, es rapido, facil, poco codigo, una sola figura, es dificil de personalizar y acceder a ciertos detalles

* El otro es el **Orientado a Objetos**, es mas dificil de usar, usa mas codigo, es mas amigable con la personalizacion, y multiples diagramas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
# grafica con el metodo PYPLOT
plt.plot(x, y, 'r')
```

```python
# gradica con el metodo ORIENTADO A OBJETOS

# figura para la grafica
fig = plt.figure()

# se crea un AXES (ejes), parametros: [PosX, PosY, SizeX, SizeY]
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5])
axes.plot(x, y, 'b')

plt.show()
```

```python
# la ventaja de usar Orientado a Objetos, es que se puede agregar mas ejes independientes
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.7])
axes2 = fig.add_axes([0.2, 0.4, 0.4, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
axes2.set_facecolor('gray')

plt.show()
```

# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = np.sin(x)
y
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x),'b')
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
fig.tight_layout()
```

# Leyendas, etiquetas, titulos, tamaño

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# PLOT NORMAL
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right')
```

```python
# PLOTS ORIENTADO A OBJETOS

fig, axes = plt.subplots(1,2,figsize=(10,5))

axes[0].plot(x,y, label='sin(x)')
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion X - Y')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```

# Colores y estilos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
# lista lista de estilos predefinicos
print(plt.style.available)
```

```python
#plt.style.use('fivethirtyeight')
#plt.style.use('clasic')
plt.style.use('bmh')
```

```python
fig, ax = plt.subplots(figsize=(5,5))

# personalizacion de lineas con estilos Matlab ('r--')
ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
```

```python
fig, ax = plt.subplots(figsize=(5,5))

# personalizacion de lines con sin estilos Mathlab

# linewidth (grosor de linea)
# linestyle (estilo de linea)
# marker (marca de cada punto)
# markersize (tamaño de marca)
# markerfacecolor (color de marca)
ax.plot(x, x+1, color='green', linewidth=2, linestyle='-', marker='o', markersize=5)
ax.plot(x, x+2, color='blue', linestyle='--', marker='x')
ax.plot(x, x+3, color='#66ff66', linestyle='dashed', marker='v', markerfacecolor='b')
ax.plot(x, x+4, color='black', alpha=0.5, linestyle=':', marker='p', markersize=10, markerfacecolor='r')
```

# Bar Plot

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
country = ['INDIA', 'JAPON', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.figure(figsize=(7,3))
plt.bar(country, population, width=0.3, color=['red','blue','green','black','yellow'])
plt.title('Poblacion por Pais')
plt.xlabel('Paises')
plt.ylabel('Poblacion')
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```

# Crear otro tipo de graficas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
data
```

```python
# histograma, analisis de distribucion y frecuencia
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

```python
# sentido vertical
plt.boxplot(data, vert=False)
plt.show()
```

```python
# patch_artist (relleno)
# notch (estrechamiento por la media)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
# se agrega un outlayer
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
# showfliers (ignora los outlayers)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```
