```python
import pandas as pd
```

```python
df = pd.read_csv('file_csv.csv')
df.head()
```

```python
def two_times(value):
    return value * 2
```

```python
df['User Rating 2'] = df['User Rating'].apply(two_times)
df.head()
```

```python
# SE PUEDE APLICAR CUALQUIER TIPO DE FUNCION
df['User Rating 2'] = df['User Rating'].apply(lambda x : x * 3)
df.head()
```

```python
df['User Rating 2'] = df.apply(lambda x : x['User Rating'] * 2 if x['Genre']=='Fiction' else x['User Rating'], axis=1)
df.head()
```
