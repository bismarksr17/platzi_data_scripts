```python
import pandas as pd
```

```python
# LECTURA DEL CSV
df = pd.read_csv('file_csv.csv')
df.head(5)
```

## FILTRADO NORMAL

```python
# FILTRADO POR SLICING A NUVEL DE FILAS
df[0:3]
```

```python
# FILTRADO DE COLUMNA, RETORNA UN NUMPY
df['Name']
```

```python
# FILTRADO DE VARIAS COLUMNAS, RETORNA UN DATAFRAME
df[['Name', 'Price']]
```

## FILTRADO USANDO LOC - ILOC

loc => filtrado por label

iloc => filtrado por indice

```python
# SIN FILTRO, SELECCIONA TODAS LAS FILAS
df.loc[:]
```

```python
# SLICING A NIVEL DE FILAS, SELECCIONA LAS FILAS EN EL RANGO 1 a 3
df.loc[1:3]
```

```python
# SLICING A NIVEL DE FILAS, SELECCIONA LAS FILAS EN EL RANGO 1 a 3
# TAMBIEN FILTRA COLUMNAS
df.loc[1:3, ['Name', 'Price']]
```

```python
# SELECCIONA TODOS LOS VALORES DE UNA COLUMNA NUMERICA Y LO MULTIPLICA POR -1
df.loc[:, ['Price']] * -1
```

```python
# SELECCIONA TODOS LOS VALORES DE LA COLUMNA INDICADA QUE CUMPLAN CON LA CONDICION (>10)
df.loc[:,['Price']] > 10
```

```python
# SELCCIONA TODAS LAS FILAS Y COLUMNAS
df.iloc[:]
```

```python
# SELECCIONA VALOR EN FILA Y COLUMNA INDICADA, Y LO MULTIPLICA POR -1
df.iloc[0,4] * -1
```

```python
# SLICING INDICANDO RANGOS DE FILAS Y COLUMASN DESEADAS
df.iloc[2:10, 4:7]
```
