```python
import pandas as pd
```

```python
df = pd.read_csv('file_csv.csv')
df.head()
```

```python
# AGRUPA POR LA COLUMNA AUTOR Y SUMA LAS COLUMNAS NUMERICAS
# NOTAR: EL LABES 'Author' ESTA UN POCO MAS ABAJO DE LOS LABES DE LAS COLUMNAS,
# ESTO ES PORQUE EL LABEL 'Author' ES UN INDEX
df.groupby('Author').sum()
```

```python
# CON LOC SE PUEDE ACCEDER AL INDICE DE AUTO INDICADO
df.groupby('Author').sum().loc['Abraham Verghese']
```

```python
# RESETEA EL INDEX, AUTHOR DEJA DE SER INDEX
df.groupby('Author').sum().reset_index()
```

```python
# MINIMO Y MAXIMO PRO COLUMNA, EN COLUMNAS NO NUMERICAS, SE REALIZA A NIVEL DE BIT
df.groupby('Author').agg(['min', 'max'])
```

```python
# ES MEJOR SELECCIONAR LAS COLUMNAS Y TIPO DE CALCULO QUE SE DESEA
df.groupby('Author').agg({'Reviews':['min', 'max'], 'User Rating':'sum'})
```

```python
# AGRUPACION CON MAS DE UNA COLUMNA, ESTO GENERA DOBLE INDEX
df.groupby(['Author','Year']).count()
```
