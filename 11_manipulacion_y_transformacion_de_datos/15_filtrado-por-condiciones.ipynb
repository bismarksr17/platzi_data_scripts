```python
import pandas as pd
```

```python
df = pd.read_csv('file_csv.csv')
df.head()
```

```python
# MASCARA BOOLEANA CON UNA DE LAS COLUMNAS DEL DF
mayor_2016 = df['Year'] > 2016
mayor_2016
```

```python
# APLICA MASCARA BOOLEANA AL DF, SE CREA UN FILTRO
df[mayor_2016]
```

```python
# TAMBIEN SE PUEDE CREAR DIRECTAMENTE LA MASCATA BOOLEANA COMO INDEX
df[df['Year'] > 2016]
```

```python
# MASCARA BOOLEANA POR IGUALDAD
fiction = df['Genre'] == 'Fiction'
fiction
```

```python
# FILTRADO POR DOBLE CONDICIONAL (AND - &)
df[fiction & mayor_2016]
```

```python
# NEGACION DE LA MASCARA BOOLEANA
df[~mayor_2016]
```

```python
# MASCARA CREADA CON STR.CONTAINS
apellido = df['Author'].str.contains('Palacio')
df[apellido]
```

```python
# MASCATA DE AÑOS EN EL RANGO DE VALORES DE 2016 A 2017
rango = df['Year'].isin([2016,2017])
df[rango]
```
