```python
import pandas as pd
import numpy as np
```

## Concat

```python
# SE CREAN 2 DF NUEVOS
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']})
df1
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']})
df2
```

```python
# SE CONCATENA, POR DEFECTO AXIS 0
# LOS INDICES HEREDAN DE LOS ANTERIOS DATAFRAMES
pd.concat([df1, df2])
```

```python
# REESTRUCTURA EL INDICE
pd.concat([df1, df2], ignore_index=True)
```

```python
# AXIS = 1, CONCATENA A NIVEL DE CPLUMNAS
pd.concat([df1, df2], axis=1)
```

## Merge

```python
# SE CREA DOS DF, CON COLUMNA EN AMBOS DF
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
print('izq:\n', izq, '\n')
print('der:\n', der)
```

```python
# MERGE INDICANDO COLUMNA A CONSIDERAR PARA COMPARAR LOS DFs
# AMBOS DF TIENE LA MISMA COLUMNA key
izq.merge(der, on='key')
```

```python
# SE CAMBIA EL NOMBRE DE LA COLUMNA key DEL SEGUNDO DF
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
print('izq:\n', izq, '\n')
print('der:\n', der)
```

```python
# EN MERGE SE TIENE QUE INDICAR LAS COLUMNAS EN AMBOS DFs A UNIR
izq.merge(der, left_on='key', right_on='key_2')
```

```python
# EN EL DF der, SE CAMBIA EL ULTIMO key A NaN
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
print('izq:\n', izq, '\n')
print('der:\n', der)
```

```python
# HOW INNER - INTERSECCION DE AMBOS DF
izq.merge(der, left_on='key', right_on='key_2', how='inner')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='outer')
```
